# NAVER WEBTOON

특정 에피소드의 베스트댓글을 읽어오기

```python
import json
import re
import requests

def get_best_comment_list(webtoon_id, ep_id):
    referer_url = 'http://comic.naver.com/webtoon/weekday.nhn'

    url = 'http://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json'
    params = {
        'ticket': 'comic',
        'templateId': 'webtoon',
        'pool': 'cbox3',
        '_callback': '_',
        'lang': 'ko',
        'country': 'KR',
        'objectId': '{}_{}'.format(webtoon_id, ep_id),
        'pageSize': 10,
        'indexSize': 10,
        'page': 1,
        'sort': 'best',
        'categoryId': '',
        'listType': 'OBJECT',
    }

    jsonp_string = requests.get(url, params=params, headers={'Referer': referer_url}).text
    json_string = jsonp_string.replace('_(', '').replace(');', '')
    
    resp = json.loads(json_string)
    return resp['result']['commentList']
```

```python
#슈퍼시크릿, 116화
#http://comic.naver.com/webtoon/list.nhn?titleId=650304&weekday=tue

best_comment_list = get_best_comment_list('650304','116')

for idx, comment in enumerate(best_comment_list,1):
    if comment['best']:
        print(idx,'''{maskedUserName}(+{sympathyCount},-{antipathyCount})
{contents}
(at {modTime})
'''.format(**comment))
```

## 한 에피소드의 이미지를 하나씩 다 받아보기

```python
#http://comic.naver.com/webtoon/detail.nhn?titleId=685460&no=3&weekday=fri

import os
import requests
from bs4 import BeautifulSoup

ep_url = 'http://comic.naver.com/webtoon/detail.nhn?titleId=20853&no=1100&weekday=tue'

html = requests.get(ep_url).text
soup = BeautifulSoup(html, 'html.parser')

for tag in soup.select('.wt_viewer img'):
    img_url = tag['src']
    img_data = requests.get(img_url, headers={'Referer': ep_url}).content
    img_name = os.path.basename(img_url)
    print(img_url)
    with open(img_name, 'wb') as f:
        f.write(img_data)
    
```

# 국민대 식단 크롤링

```python
import json
import requests

params = {
    'callback': '_',
    'sdate': '2017-08-01',
    'edate': '2017-08-31',
}

menu_url = 'http://kmucoop.kookmin.ac.kr/menu/menujson.php'

jsonp_string = requests.get(menu_url, params=params).text
json_string = jsonp_string.replace('_(', '').replace(');', '')
menu_dict = json.loads(json_string)
menu_dict
```

```python
menu_dict.keys()
```

```python
for date, category_list in menu_dict['한울식당'].items():
    print(date)
    for name, menu in category_list.items():
        if menu['메뉴']:
            menu_names = ' '.join(menu['메뉴'].split())
            print(name, menu_names, menu['가격'])
    print()
```
